```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("shopping_trends.csv")

# Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Gender Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution of Customers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("shopping_trends.csv")

# Plot the violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='Gender', y='Age', data=data)
plt.title('Age and Gender Distribution of Customers')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("shopping_trends.csv")

# Calculate the distribution of product categories
category_counts = data['Category'].value_counts()

# Plot the pie chart
plt.figure(figsize=(10, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Product Category Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show
```


1. **Popular Product Categories**: The chart highlights the most popular product categories based on the proportion of total purchases. For example, if "Clothing" occupies the largest segment, it indicates that clothing items are the most frequently purchased products.

2. **Least Popular Categories**: Similarly, the chart shows the least popular product categories. If "Accessories" has the smallest segment, it suggests that accessories are less frequently purchased compared to other categories.

3. **Market Demand**: The distribution of product categories provides insights into market demand. Categories with larger segments indicate higher demand, which can help businesses focus their inventory and marketing efforts on these products.

4. **Customer Preferences**: The chart can reveal customer preferences for different types of products. Understanding these preferences can help businesses tailor their product offerings to better meet customer needs.

5. **Sales Strategy**: The distribution can inform sales strategies. For example, if certain categories are underperforming, businesses might consider promotions or discounts to boost sales in those areas.

6. **Inventory Management**: Knowing which categories are most and least popular can help with inventory management. Businesses can ensure they have sufficient stock of high-demand items and avoid overstocking less popular products.

```python
# Impact of Discounts
plt.figure(figsize=(10, 6))
sns.boxplot(x='Discount Applied', y='Purchase Amount (USD)', data=data)
plt.title('Impact of Discounts on Purchase Amount')
plt.xlabel('Discount Applied')
plt.ylabel('Purchase Amount (USD)')
plt.show()

# Impact of Promo Codes
plt.figure(figsize=(10, 6))
sns.boxplot(x='Promo Code Used', y='Purchase Amount (USD)', data=data)
plt.title('Impact of Promo Codes on Purchase Amount')
plt.xlabel('Promo Code Used')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```


### Impact of Discounts
- The box plot for discounts shows the distribution of purchase amounts for transactions where discounts were applied versus those where they were not.
- If the median purchase amount is higher for transactions with discounts, it suggests that discounts encourage customers to spend more.
- The spread of the data (interquartile range) can indicate the variability in purchase amounts. A wider spread for discounted purchases might suggest that discounts attract a broader range of spending behaviors.

### Impact of Promo Codes
- The box plot for promo codes shows the distribution of purchase amounts for transactions where promo codes were used versus those where they were not.
- Similar to discounts, if the median purchase amount is higher for transactions with promo codes, it suggests that promo codes incentivize higher spending.
- The presence of outliers in the box plot can indicate exceptional cases where promo codes led to significantly higher or lower purchase amounts.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("shopping_trends.csv")

# Calculate the max and min values
location_counts = data['Location'].value_counts()
max_location = location_counts.idxmax()
max_count = location_counts.max()
min_location = location_counts.idxmin()
min_count = location_counts.min()

# Plot the geographical analysis with larger labels
plt.figure(figsize=(20, 18))
sns.countplot(y='Location', data=data, order=location_counts.index)
plt.title('Sales by Location', fontsize=20)
plt.xlabel('Count', fontsize=15)
plt.ylabel('Location', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

# Print the max and min values
print(f"Location with the maximum sales: {max_location} ({max_count} sales)")
print(f"Location with the minimum sales: {min_location} ({min_count} sales)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("shopping_trends.csv")

# Group the data by location and calculate the total purchase amount
location_sales = data.groupby('Location')['Purchase Amount (USD)'].sum().sort_values(ascending=False)

# Create a DataFrame for the heatmap
heatmap_data = pd.DataFrame(location_sales).reset_index()

# Plot the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data.pivot_table(index='Location', values='Purchase Amount (USD)'), annot=True, fmt='.2f', cmap='YlGnBu')
plt.title('Heatmap of Total Purchase Amount by Location')
plt.xlabel('Total Purchase Amount (USD)')
plt.ylabel('Location')
plt.show()
```

1. **Top Performing Locations**: The locations with the highest sales can be identified. These locations are crucial for the business as they contribute significantly to the overall sales. For example, if California has the highest count, it indicates a strong customer base there.

2. **Underperforming Locations**: Similarly, the locations with the lowest sales can be identified. These areas might need targeted marketing efforts or further investigation to understand the low performance.

3. **Regional Trends**: The graph can reveal regional trends in customer behavior. For instance, certain regions might prefer specific products or have higher purchase frequencies.

4. **Resource Allocation**: Understanding which locations perform well and which do not can help in better resource allocation. More resources can be directed towards high-performing areas to maximize sales, while strategies can be developed to improve sales in low-performing areas.

5. **Market Expansion**: The data can also help in identifying potential markets for expansion. Locations with moderate sales but high potential can be targeted for growth.

6. **Customer Preferences**: The geographical distribution can provide insights into customer preferences in different regions, helping tailor marketing strategies to suit regional tastes and preferences.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("shopping_trends.csv")

# Plot the box plot for sentiment analysis of reviews by item
plt.figure(figsize=(14, 8))
sns.boxplot(x='Item Purchased', y='Review Rating', data=data)
plt.title('Sentiment Analysis of Reviews by Individual Items')
plt.xlabel('Item Purchased')
plt.ylabel('Review Rating')
plt.xticks(rotation=45)
plt.show()
```

1. **Customer Satisfaction**: The distribution of review ratings for each item provides insights into customer satisfaction. Items with higher median ratings indicate higher customer satisfaction, while items with lower median ratings may need improvement.

2. **Consistency in Quality**: The spread of the ratings for each item can indicate the consistency in quality. A narrower spread suggests consistent quality, while a wider spread indicates variability in customer experiences.

3. **Outliers**: The presence of outliers in the box plot can highlight exceptional cases where customers had significantly better or worse experiences with certain items. These outliers can provide valuable feedback for product improvement.

4. **Popular Items**: Items with a higher number of reviews can be identified as popular products. These items are likely to have a larger customer base and can be prioritized for inventory and marketing efforts.

5. **Areas for Improvement**: Items with lower median ratings or a higher number of negative reviews can be identified as areas for improvement. Businesses can focus on addressing the issues with these items to enhance customer satisfaction.

6. **Product Performance Comparison**: By comparing the review ratings of different items, businesses can identify which products are performing well and which ones are not. This comparison can help in making data-driven decisions for product development and marketing strategies.
